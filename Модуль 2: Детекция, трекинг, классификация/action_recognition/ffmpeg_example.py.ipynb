**Извлекаем информацию о видео при помощи ffprobe**

```python
video_path = 'examples/test.mp4'

!ffprobe -loglevel 0 -print_format json -show_format -show_streams {video_path}
```

**Извлечение фреймов из видео**

```python
video_path = 'examples/test.mp4'

! ffmpeg -i {video_path} -vf fps=30 -ss 00:00:01 -t 00:00:3 -vsync 0 -qscale:v 2 examples/frames/img-%6d.jpeg
```

**Извлечение фреймов из видео (с изменением разрешения)**

```python
video_path = 'examples/test.mp4'

! ffmpeg -i {video_path} -vf fps=1 -s 1920x1080 -vsync 0 -qscale:v 2 examples/frames_hd/img-%6d.jpeg 
```

```python
video_path = 'examples/test.mp4'

!ffmpeg -i {video_path} -r 0.5 -qscale:v 2 examples/frames_fps0p5/img-%5d.jpeg
```

**Конвертируем фреймы в видео**

```python
frames_dir = 'examples/frames'
video_path = 'examples/video_from_frames.mp4'

! ffmpeg -pattern_type glob -framerate 30 -f image2 -i '{frames_dir}/*.jpeg' {video_path}
```

**Перекодируем видео**

```python
src = 'examples/test.mp4'
dst = 'examples/test_converted.mp4'

! ffmpeg -i {src} -codec:v libx264 {dst}
```

**Вырезаем видео-сегмент**

```python
src = 'examples/test_converted.mp4'
dst = 'examples/test_segment.mp4'

! ffmpeg -ss 00:00:05.0 -i {src} -c copy -t 00:00:02.0 {dst}
```

**Конкатенация нескольких видео в один файл**

```python
src1 = 'examples/test.mp4'
src2 = 'examples/test_converted.mp4'
dst = 'examples/concat.mp4'

! echo file {src1} >> mylist.txt 
! echo file {src2} >> mylist.txt
! ffmpeg -f concat -safe 0 -i mylist.txt -c copy {dst}
```

**Объединяем видео в вертикальный grid**

```python
src1 = 'examples/test.mp4'
src2 = 'examples/test_converted.mp4'
dst = 'examples/grid.mp4'

! ffmpeg -i {src1} -i {src2} -filter_complex vstack=inputs=2 -c:v libx264 -crf 23 {dst}
```

**Делаем GIF из видео**

```python
src = 'examples/grid.mp4'
dst_gif = 'examples/test.gif'

! ffmpeg -i {src} -r 10 -vf scale=512:-1 -ss 00:00:01 -to 00:00:3 {dst_gif}
```
